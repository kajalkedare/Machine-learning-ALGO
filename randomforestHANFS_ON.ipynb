Out of Bag (OOB)
The data that was left and was never seen by my model
We can use this data for validation

```python
y
```

```python
x
```

**We have to build a model that use oob parameter**

**Important**
* max_feature:
    * sqrt -> by default (It sqrt(n))
    * log2 --> uses log base 2(n)
    * int -> numbers (5)
    * float -> 0.5 (fraction of total feature)
    * None -> it uses all features (like normal decison)

**Note: Smaller value of max_feature: more randomness---> Less correlation between treens ---> More generalization ---> less Overfitting**

```python
oobclassifier= RandomForestClassifier(n_estimators=100,oob_score=True,max_features='sqrt',random_state=0)
oobclassifier.fit(x_train,y_train)
```

```python
print(oobclassifier.oob_score_)#78% oob score
oobError = 1-oobclassifier.oob_score_
print(oobError)
```

```python
y_pred=oobclassifier.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

SMOTE(Synthetic Minority oversampling technique)
SMOTE creates new data points for minority class by combining nearby existing ones
It helps us balancing the dataset, makes my model more identifying for minority class, which basically imporoves my recall
