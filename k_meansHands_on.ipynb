building and evaluating the model

E++++++

```python
from sklearn .cluster import KMeans
```

```python
wcss =[]
k_range = range(1,11)

for k in k_range:
    kmeans = KMeans(n_clusters=k , n_init=10, random_state = 42,init='k-means++')
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss = []
k_range = range(1,11) # we will test k from 1 to 10

for k in k_range:
  kmeans = KMeans(n_clusters=k,n_init = 10, random_state=42,init = 'k-means++')
  kmeans.fit(x_scaled)
  wcss.append(kmeans.inertia_) # 'inertia_ is the wcss
```

```python
wcss
```

```python
# plt.plot(k_range,wcss,marker="*",color='blue')
# plt.xlabel('number of cluster')
# plt.ylable('wcss(inertia)')
# plt.grid(True)
# plt.show()

plt.plot(k_range,wcss,marker='*',color='blue')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score =[]
k_range = range(2,11)
for k in k_range:
  kmeans = KMeans(n_clusters=k,n_init = 10, random_state=42,init = 'k-means++')
  kmeans.fit(x_scaled)
  score = silhouette_score(x_scaled,kmeans.labels_)
  sil_score.append(score)
```

```python
sil_score
```

```python
plt.plot(k_range,sil_score,marker='o',color='red')
plt.title('The silhoutte Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Average silhoou')
plt.grid(True)
plt.show()
```

```python
k=3
kmeans_final = KMeans(n_clusters= k,init='k-means++',n_init=10,random_state=42)
kmeans_final.fit(x_scaled)
```

```python
cluster_labels = kmeans_final.labels_
cluster_labels
```
